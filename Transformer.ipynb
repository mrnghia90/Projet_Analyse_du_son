```python
%reset -f
import os
import pandas as pd
import PIL
from PIL import Image

import librosa
import librosa.display
import numpy as np
```

```python
def audio2spectrogram(audio_file, n_mels = 128, path_read = None, path_save = None):
    #audio2spectrogram
    #audio_file:
    #n_mels:
    #path:
    
    # Lire audio
    y, sr= librosa.load(path_read + audio_file)
    
    # Scaling mel spectrogram
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels*2)
    # Transformation log
    S_db_mel = librosa.power_to_db(S, ref=np.max)
    
    # Convert en np.array
    S_array = np.array(S_db_mel).astype(np.float32)
    # Convert en Image
    S_im = Image.fromarray(S_array, mode ='F')

    image_path = os.path.join(path_save, os.path.basename(audio_file)[:-4] +'.tiff')
    
    # Enregistrement
    S_im.save(image_path)
```

```python
# Create an empty list to store the data
data = []
i = 0
# Iterate through subfolders and audio files, applying the transformation function and populating the data list
for root, subdirs, files in os.walk('data'):
    for file in files:
        if file.endswith('.wav'):  # Adjust the prefix as needed
            print('Transfrom sample : ',i)
            audio_file = os.path.join(root, file)
            audio_id = audio_file
            
            path_read = audio_file[:8]
            audio_name =audio_file[8:]
            label = audio_name[0]
            person = audio_name[2:4]
            audio_id = audio_file[5:7] +'_'+ audio_name[:-4]
            audio2spectrogram(audio_name,path_read=path_read, path_save ='data_img' )
            data.append({'id': audio_id, 'person': person, 'label': label})
            i += 1

df = pd.DataFrame(data)

df.to_csv('Metadonnee', index=False)
```
